## 3. Data Quality Check - Identify Missing Values
### Check for null values in all columns

```python
df.isnull().sum()
```

### Issue Identified: Price Mapping Inconsistency
**Problem**: `initial_price`, `discount`, and `final_price` columns are not properly aligned  
**Observation**: The relationship `final_price = initial_price - (initial_price × discount%)` is not maintained

```python
df[['initial_price','discount','final_price']]
```

## 4. Handle Missing Values
**Strategy**:
- **Numeric columns** (float64, int64): Fill with `0`
- **Categorical columns** (object): Fill with `"Unknown"`

```python

for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col] = df[col].fillna(0)
    else:
        df[col] = df[col].fillna("Unknown")
```

### Verification: Confirm All Missing Values Have Been Filled
✓ Validating null value count after filling

```python
df.isnull().sum()
```

## 5. Correct Price Calculation
**Formula**: `final_price = initial_price - (initial_price × discount / 100)`  
This ensures accurate pricing after discount is applied

```python
df['final_price'] = ((df['initial_price']-(df['initial_price']*df['discount']/100)))
```

```python
df[['initial_price','discount','final_price']]
```

## 6. Data Filtering & Feature Selection - Identify Popular Products
**Filter Criteria**: Select products with `rating > 4.0`  
**Selected Columns**: product_id, title, description, rating, ratings_count, final_price, currency, category

```python
PopularProduct = df[df['rating'] > 4][['product_id','title','product_description','rating','ratings_count','final_price','currency','category']]
```

## 7. Remove Duplicate Records
**Purpose**: Eliminate duplicate rows to ensure data uniqueness and integrity

```python
df = df.drop_duplicates()
```

## 8. Feature Engineering - Calculate Total Sales Amount
**New Feature**: `total_amount = final_price × ratings_count`  
**Purpose**: Represents the total revenue impact of each product based on its sales volume and price

```python
df['total_amount'] = df['final_price'] * df['ratings_count']
```
